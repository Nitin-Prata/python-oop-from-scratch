```python
class Atm:

    # constructor 
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input("""
        Hi how can I help you?
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anythig else to exit
        """)

        if user_input == '1':
            self.create_pin()
        # create pin
        elif user_input == '2':
            self.change_pin()
        #change pin
        elif user_input == '3':
            self.check_balance()
        # check balance
        elif user_input == '4':
            self.withdraw()
        # withdraw
        else:
            exit() 

    def create_pin(self):
        user_pin = input('enter your pin')
        self.pin = user_pin

        user_balance = int(input('enter balance'))
        self.balance = user_balance

        print('pin created successfuly')
        self.menu()

    def change_pin(self):
        old_pin = input('enter old pin')

        if old_pin == self.pin:
            new_pin = input('enter new pin')
            self.pin = new_pin
            print('pin change successful')
            self.menu()
            # let him change the pin

        else:
            print('can not change')
            self.menu()

    def check_balance(self):
        user_pin = input('enter your pin')
        
        if user_pin == self.pin:
            print('your balance is',self.balance)
            self.menu()

        else:
            print('pin is not correct')
            self.menu()
            
    
    def withdraw(self):
        user_pin = input('enter the pin')

        if user_pin == self.pin:
            # allow to withdraw
            amount = int(input('enter the amount'))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('withdrawl successful',self.balance)
            else:
                print('u are so poor')
        else:
            print('sale chor')
        self.menu()
```
